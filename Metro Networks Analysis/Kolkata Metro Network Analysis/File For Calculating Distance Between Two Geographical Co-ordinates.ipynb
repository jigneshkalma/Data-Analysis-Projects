```python
# This Code is Example For How I Calculate Distance Between Two Geographical Co-ordinates
# First install geopy library using this command : "!pip install geopy"
from geopy.distance import geodesic

# Define coordinates
coords_1 = (52.2296756, 21.0122287)
coords_2 = (51.919438, 19.145136)

# Calculate distance
distance = geodesic(coords_1, coords_2).km

print(distance)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("Kolkata-Metro-Network.csv")
```

```python
station_1_coords = (df.loc[df['Station ID'] == 1, 'Latitude'].values[0], df.loc[df['Station ID'] == 1, 'Longitude'].values[0])
print(station_1_coords)
```

```python
for index, row in df.loc[df['Station ID'].between(2, 26)].iterrows():
    station_n_coords = (row['Latitude'], row['Longitude'])
    distance = geodesic(station_1_coords, station_n_coords).kilometers
    df.loc[index, 'Distance from Start (km)'] = round(distance, 1)

df
```
